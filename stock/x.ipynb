```python
import matplotlib.pyplot as plt

import numpy as np
import tushare as ts
import pandas as pd

from mpl_toolkits.mplot3d import Axes3D  #D
```

```python
#  PEPB 
ts.set_token('42731ca565c5d019007ef5cd7db7808757b2ea3fdbfb31d4f7b61444b')
pro = ts.pro_api()

df_base = pro.stock_basic() # ts.get_stock_basics()
df_base['code'] = df_base.index

# 
df_todays = ts.get_today_all()
df_todays['code'] = df_todays.index

df_roe = ts.get_report_data(2018,3)
df_roe['code'] = df_roe.index
```

```python
#  PEPB 
df = pd.merge(df_todays, df_base, how='left', on=['code'])
df = pd.merge(df, df_roe, how='left', on=['code'])

df.dropna()

result = df[(0<df['pb'])&(df['pb']<8)
             &(-20<df['roe'])&(df['roe']<30)
             &(0<df['trade'])&(df['trade']<70)]

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

X = result['pb'].tolist()  # X  PB   
Y = result['roe'].tolist()
Z = result['trade'].tolist()  # Z   

ax.set_xlabel('PB', color='g')
ax.set_ylabel('roe', color='g')
ax.set_zlabel('trade', color='b')#

ax.scatter(X, Y, Z)

plt.show()
```
